```python
import numpy as np
from numpy.linalg import det
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create time series as first Barret example

np.random.seed(100)
time = np.arange(0,5,0.01)

alpha = 0.2
beta = 0.5
data = np.zeros((2,len(time)))

for i in range(1,len(time)):
    data[0,i] = alpha*data[0,i-1] + beta*data[1,i-1] + np.random.normal()
    data[1,i] = np.random.normal()
```

```python
plt.plot(time,data,label="x")
plt.plot(time,data,label="y")
plt.xlabel("t")
plt.legend()
plt.show()
```

```python
# slice for PID

Y = data[0,:-1]
Z = data[1,:-1]
X = data[:,1:]
```

```python
def PID(X,Y,Z): # X is the target, Y and Z are the starting points
    
    # det of all partial matrices
    sig_x = det(np.cov(X))
    sig_y = np.var(Y)
    sig_z = np.var(Z)
    sig_xy = det(np.cov(np.vstack((X,Y))))
    sig_xz = det(np.cov(np.vstack((X,Z))))
    sig_yz = det(np.cov(np.vstack((Y,Z))))
    sig_xyz = det(np.cov(np.vstack((X,Y,Z))))
    
    # mutual info
    I_xy = np.log2(sig_x*sig_y/sig_xy)
    I_xz = np.log2(sig_x*sig_z/sig_xz)
    I_xyz = np.log2(sig_x*sig_yz/sig_xyz)
    
    # PID starting from R
    R = min(I_xy,I_xz)
    
    if R == I_xy:
        Uy = 0
        Uz = I_xz - I_xy
        S = I_xyz - I_xz   
    else:
        Uz = 0
        Uy = I_xy - I_xz
        S = I_xyz - I_xy
    
    return Uy, Uz, S, R
```

```python
Uy, Uz, S, R = PID(X,Y,Z)
print("Uy:",Uy)
print("Uz:",Uz)
print("S:",S)
print("R",R)
```
